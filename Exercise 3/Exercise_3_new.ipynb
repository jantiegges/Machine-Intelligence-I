```python
# load own package
os.chdir("/Users/jantiegges/Google Drive/Uni/1. Semester/Machine Intelligence I/Machine-Intelligence-I/ML_ressources")

from multilayer_perceptron import MultilayerPerceptron
```

```python
X = data[:, 0]
y = data[:, 1]

X = np.expand_dims(X, axis=1)
y = np.expand_dims(y, axis=1)

sizes = [1, 3, 1]
mlp = MultilayerPerceptron(sizes, init_limit=0.5, learning_rate=0.05)
error = mlp.fit(X, y)
print(f"Final Error: {error[-1]}")
```

```python
plt.plot(error)
plt.ylabel("Mean squared error")
plt.xlabel("Iterations")
plt.title("First initialization error")
plt.show()
```
